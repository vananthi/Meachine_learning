**X & Y Split :**

*   X is a matrix of feature values. It is also known as input, feature set, or independent variables.

*   Y is a vector of target values. It is also known as target, class, or label.

```python
#X,Y Split
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
#Data Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
std_data = scaler.transform(X)
```

```python
std_data
```

```python
X = std_data
```

```python
X
```

```python
Y
```

**Train & Test Split :**

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state= 2)
```

```python
(X.shape, X_train.shape, X_test.shape)
```

```python
(Y.shape, Y_train.shape, Y_test.shape)
```

**Model Building :**

**Support Vector Mechine : ( SVM - Classifier )**

*   
It is used to Supervised Learning models to solve Classification, Regression and Outlier detection problems.
*   It is a Binary Classifier.
image.png



```python
#Model Building
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, Y_train)
```

```python
Y_predict = svc.predict(X_test)
```

**Model Evaluation :**

*   Evaluation metrics are quantitative measures that assess the performance and effectiveness of a statistical or machine learning model.
*   
They provide insights into how well the model is performing and help in comparing different models or algorithms.

**Confusion Matrix :**


*   A Confusion matrix is an N x N matrix used for evaluating the performance of a classification model, where N is the total number of target classes.
*   The matrix compares the actual target values with those predicted by the machine learning model.

image.png





```python
#Model Evaluation
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_predict)
score_svc = svc.score(X_test, Y_test)
```

```python
cm
```

```python
score_svc
```

```python
score_svctr = svc.score(X_train,Y_train)
```

```python
score_svctr
```

**DECISION TREE ALGORITHM:**

Decision trees are a type of machine-learning algorithm that can be used for both classification and regression tasks. They work by learning simple decision rules inferred from the data features. These rules can then be used to predict the value of the target variable for new data samples.

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=2)
dtc.fit(X_train,Y_train)
```

```python
Y_predict = dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_predict)
score_dtc = dtc.score(X_test,Y_test)
```

```python
cm
```

```python
score_dtc
```

```python
score_dtctr = dtc.score(X_train,Y_train)
```
